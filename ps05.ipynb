### Problem 1

 - How many passengers are recorded in this log?
 - What percentage are male?
 - What percentage survived?

Hint: You can find these three answers with a single command.

````python
import pandas as pd
import matplotlib.pyplot as plt

pp_df = pd.read_csv('titanic.csv')

total_passengers = len(pp_df)

percent_male = (pp_df["Sex"].sum() / total_passengers) * 100

percent_survived = (pp_df["Survived"].sum() / total_passengers) * 100

print("Total passengers:", total_passengers)
print("Percentage of males:", percent_male, "%")
print("Percentage of survivors:", percent_survived, "%")
````

### Problem 2:

 - What percentage of female passengers survived?

````python
import pandas as pd
import matplotlib.pyplot as plt

pp_df = pd.read_csv('titanic.csv')

total_females = len(pp_df[pp_df["Sex"] == 0])
total_males = len(pp_df[pp_df["Sex"] == 1])

female_survivors = len(pp_df[(pp_df["Sex"] == 0) & (pp_df["Survived"] == 1)])
male_survivors = len(pp_df[(pp_df["Sex"] == 1) & (pp_df["Survived"] == 1)])

percent_female_survived = (female_survivors / total_females) * 100 if total_females > 0 else 0
percent_male_survived = (male_survivors / total_males) * 100 if total_males > 0 else 0

print("Percentage of female passengers who survived:", percent_female_survived, "%")
print("Percentage of male passengers who survived:", percent_male_survived, "%")
````

### Problem 3

Use ```groupby``` and ```agg``` to find the percentage of the total tickets and the survivorship percentage for each class of ticket.

````python
import pandas as pd
import matplotlib.pyplot as plt

pp_df = pd.read_csv('titanic.csv')

# Class 1
grouped = pp_df.groupby("Pclass").agg(
    total_tickets=("PassengerId", "count"),  
    total_survivors=("Survived", "sum"),    
)

total_tickets_all = grouped["total_tickets"].sum()  
class_1_tickets = grouped.loc[1, "total_tickets"]   
class_1_survivors = grouped.loc[1, "total_survivors"]  

percentage_tickets_class_1 = (class_1_tickets / total_tickets_all) * 100
survivorship_percentage_class_1 = (class_1_survivors / class_1_tickets) * 100

print("Class 1 % Tickets:", percentage_tickets_class_1, "%")
print("Class 1 % Survivorship:", survivorship_percentage_class_1, "%")

# Class 2
grouped = pp_df.groupby("Pclass").agg(
    total_tickets=("PassengerId", "count"),  
    total_survivors=("Survived", "sum"),    
)

total_tickets_all = grouped["total_tickets"].sum()  
class_2_tickets = grouped.loc[2, "total_tickets"]   
class_2_survivors = grouped.loc[2, "total_survivors"]  

percentage_tickets_class_2 = (class_2_tickets / total_tickets_all) * 100
survivorship_percentage_class_2 = (class_2_survivors / class_2_tickets) * 100

print("Class 2 % Tickets:", percentage_tickets_class_2, "%")
print("Class 2 % Survivorship:", survivorship_percentage_class_2, "%")

# Class 3
grouped = pp_df.groupby("Pclass").agg(
    total_tickets=("PassengerId", "count"),  
    total_survivors=("Survived", "sum"),    
)

total_tickets_all = grouped["total_tickets"].sum()  
class_3_tickets = grouped.loc[3, "total_tickets"]   
class_3_survivors = grouped.loc[3, "total_survivors"]  

percentage_tickets_class_3 = (class_3_tickets / total_tickets_all) * 100
survivorship_percentage_class_3 = (class_3_survivors / class_3_tickets) * 100

print("Class 3 % Tickets:", percentage_tickets_class_3, "%")
print("Class 3 % Survivorship:", survivorship_percentage_class_3, "%")
````

### Problem 4

Create two new dataframes corresponding to survivors and non-survivors keeping only the columns: age, name, and sex. Display the first five rows of these new dataframes.

Create a figure showing the distribution of age by sex and survivorship (four histograms on the same axes: male-survivors, female-survivors, male-unlucky, female-unlucky). Choose colors and histtype (see the [documentation for histogram](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)) to make this figure easy to read and interpret.

````python
import pandas as pd
import matplotlib.pyplot as plt

pp_df = pd.read_csv('titanic.csv')

survivors_df = pp_df[pp_df["Survived"] == 1][["Age", "Name", "Sex"]]
non_survivors_df = pp_df[pp_df["Survived"] == 0][["Age", "Name", "Sex"]]

print("Survivors DataFrame:")
print(survivors_df.head())

print("\nNon-Survivors DataFrame:")
print(non_survivors_df.head())

plt.figure(figsize=(10, 6))

plt.hist(
    survivors_df[survivors_df["Sex"] == 1]["Age"].dropna(),
    alpha=0.7,
    label="Male Survivors",
    color="blue",
    histtype="stepfilled",
)
plt.hist(
    survivors_df[survivors_df["Sex"] == 0]["Age"].dropna(),
    alpha=0.7,
    label="Female Survivors",
    color="green",
    histtype="stepfilled",
)
plt.hist(
    non_survivors_df[non_survivors_df["Sex"] == 1]["Age"].dropna(),
    alpha=0.5,
    label="Male Unlucky",
    color="red",
    histtype="stepfilled",
)
plt.hist(
    non_survivors_df[non_survivors_df["Sex"] == 0]["Age"].dropna(),
    alpha=0.5,
    label="Female Unlucky",
    color="orange",
    histtype="stepfilled",
)

plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution by Sex and Survivorship")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
pp_df = pd.read_csv('titanic.csv')
pp_df.head()
````

````python
label = ['Survived']
features = ['Sex', 'Age', 'Pclass']

titanic_df = pd.read_csv('titanic.csv') 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(titanic_df[features], titanic_df[label], test_size = 0.2)
````

````python
from sklearn.tree import DecisionTreeClassifier , plot_tree

tree_clf = DecisionTreeClassifier(max_depth = 4)
tree_clf.fit(X_train, y_train)

fig, ax = plt.subplots(1,1, figsize = (15,6))
plot_tree(tree_clf, feature_names = features, class_names = ['Drowned', 'Survived'],
          filled = True, fontsize = 6)
plt.show()
````

````python
from sklearn.metrics import ConfusionMatrixDisplay

y_tree_train = tree_clf.predict(X_train)
y_tree_test = tree_clf.predict(X_test)

fig, ax = plt.subplots(1,2, figsize = (10, 5))

ConfusionMatrixDisplay.from_predictions(y_train, y_tree_train, 
                                        normalize = 'true',
                                        display_labels = ['Drowned', 'Survived'],
                                        ax = ax[0])

ConfusionMatrixDisplay.from_predictions(y_train, y_tree_train, 
                                        normalize = 'true',
                                        display_labels = ['Drowned', 'Survived'], 
                                        ax = ax[1])

ax[0].set_title('training')
ax[1].set_title('testing')
fig.suptitle('Decision Tree')
plt.show()
````
